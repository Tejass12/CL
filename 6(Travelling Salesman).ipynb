```python
import numpy as np
import random

# Distance matrix
distances = np.array([
    [0, 2, 9, 10, 7],
    [2, 0, 6, 4, 3],
    [9, 6, 0, 8, 5],
    [10, 4, 8, 0, 6],
    [7, 3, 5, 6, 0]
])

num_cities = len(distances)
num_ants = 5
num_iterations = 50

alpha = 1          # Pheromone importance
beta = 2           # Distance importance
evaporation = 0.5
Q = 100            # Pheromone deposit factor

# Initialize pheromone matrix
pheromone = np.ones((num_cities, num_cities))


# Function to calculate route length
def route_length(route):
    length = 0
    for i in range(len(route) - 1):
        length += distances[route[i]][route[i + 1]]
    # Return to starting city
    length += distances[route[-1]][route[0]]
    return length
# Function to choose next city
def choose_next_city(current, visited):
    probabilities = []
    for j in range(num_cities):
        if j not in visited:
            tau = pheromone[current][j] ** alpha
            eta = (1 / distances[current][j]) ** beta
            probabilities.append((j, tau * eta))

    total = sum(p for city, p in probabilities)
    probabilities = [
        (city, p / total)
        for city, p in probabilities
    ]
    r = random.random()
    cumulative = 0

    for city, prob in probabilities:
        cumulative += prob
        if r <= cumulative:
            return city

# Store best route
best_route = None
best_length = float('inf')

# Main ACO loop
for iteration in range(num_iterations):
    all_routes = []
    for ant in range(num_ants):
        start = random.randint(0, num_cities - 1)
        route = [start]
        while len(route) < num_cities:
            next_city = choose_next_city(route[-1], route)
            route.append(next_city)
        length = route_length(route)
        all_routes.append((route, length))
        # Update best route
        if length < best_length:
            best_length = length
            best_route = route

    # Evaporation
    pheromone *= (1 - evaporation)
    # Pheromone update
    for route, length in all_routes:
        for i in range(len(route) - 1):
            pheromone[route[i]][route[i + 1]] += Q / length
        pheromone[route[-1]][route[0]] += Q / length
    print(f"Iteration {iteration + 1}: Best Length = {best_length}")

# Final output
print("\nBest Route:", best_route)
print("Shortest Distance:", best_length)
```
