```python
import numpy as np
import random

# Dummy dataset (features: vibration, amplitude)
X = np.random.rand(50, 2)

# Simple classification rule
y = np.array([0 if sum(x) < 1 else 1 for x in X])

# Distance function (similarity measure)
def distance(a, b):
    return np.linalg.norm(a - b)

# Initialize antibodies
def init_population(size):
    return [np.random.rand(2) for _ in range(size)]

# Fitness function
def fitness(ab, X, y):
    score = 0
    for i in range(len(X)):
        pred = 1 if distance(ab, X[i]) < 0.5 else 0
        if pred == y[i]:
            score += 1
    return score


# Clone antibodies
def clone(pop, factor):
    clones = []
    for p in pop:
        for _ in range(factor):
            clones.append(p.copy())
    return clones

# Mutate clones
def mutate(clones):
    for c in clones:
        c += np.random.uniform(-0.1, 0.1, size=2)
    return clones

# Selection process
def select(pop, X, y, size):
    scored = [(fitness(p, X, y), p) for p in pop]
    scored.sort(reverse=True, key=lambda x: x[0])
    return [p for score, p in scored[:size]]

# AIS Algorithm
def AIS():
    pop_size = 10
    generations = 10
    population = init_population(pop_size)
    for gen in range(generations):
        selected = select(population, X, y, pop_size // 2)
        clones = clone(selected, 3)
        mutated = mutate(clones)
        population = select(population + mutated, X, y, pop_size)
        print(f"Generation {gen + 1} completed")
    return population

# Train AIS
antibodies = AIS()

# Test classification
test = np.array([0.3, 0.4])   # Sample input
best = antibodies[0]
pred = 1 if distance(best, test) < 0.5 else 0

print("\nTest Sample:", test)
print("Prediction:", "Damaged" if pred == 1 else "Healthy")
```
