```python
import random

# Fitness function
def fitness(x):
    return x * x


# Parameters
population_size = 6
clone_count = 3
mutation_rate = 0.2
generations = 10


# Initialize population
population = [
    random.uniform(0, 10)
    for _ in range(population_size)
]


# Main loop
for gen in range(generations):

    # Sort population based on fitness
    population.sort(key=fitness, reverse=True)

    # Select best antibodies
    selected = population[:3]

    clones = []

    # Cloning and mutation
    for antibody in selected:

        for i in range(clone_count):

            mutated = antibody + random.uniform(
                -mutation_rate,
                mutation_rate
            )

            clones.append(mutated)

    # Form new population
    population = selected + clones

    population = population[:population_size]


# Best solution
best = max(population, key=fitness)

print("Best Antibody Value:", round(best, 2))

print("Best Fitness Value:", round(fitness(best), 2))
```
