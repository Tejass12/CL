```python
import xmlrpc.client

# Connect to server
proxy = xmlrpc.client.ServerProxy("http://127.0.0.1:9000/")

# Input from user
num = input("Enter a positive integer: ")

# Check input
if num.isdigit():

    num = int(num)

    # Limit large numbers
    if num > 100:
        print("Number factorial is too big to display.")

    else:
        # Call remote function
        result = proxy.factorial(num)

        # Output
        print("Factorial is:", result)

else:
    print("Invalid input! Please enter only positive integers.")
```
