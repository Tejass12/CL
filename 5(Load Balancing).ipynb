```python
import random

# List of servers
servers = ["Server-1", "Server-2", "Server-3"]

# Client requests
requests = [
    "Request-1",
    "Request-2",
    "Request-3",
    "Request-4",
    "Request-5",
    "Request-6"
]

print("Servers:", servers)

print("Client Requests:", requests)


# Round Robin Load Balancing

print("\nRound Robin Load Balancing:")

server_index = 0

for req in requests:

    print(req, "->", servers[server_index])

    server_index = (server_index + 1) % len(servers)


# Random Load Balancing

print("\nRandom Load Balancing:")

for req in requests:

    selected_server = random.choice(servers)

    print(req, "->", selected_server)
```
